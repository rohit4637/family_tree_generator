```python
from pdf2image import convert_from_path
poppler_path=r"C:\Users\welcome\Downloads\poppler-22.04.0\Library\bin"
pdf_path=r"validateUser.pdf"
image = convert_from_path(pdf_path=pdf_path,poppler_path=poppler_path)
image=image[6:7]
```

```python
#### from PIL import Image
import re
import pytesseract
from pytesseract import image_to_string

def crop(im, height, width):
    # k=1
    # im = Image.open(input)
    imgwidth, imgheight = im.size
    list=[]
    voter_list=[]
    for i in range(0,imgheight,height):
        for j in range(0,imgwidth,width):
            box = (j, i+32, j+width-140, i+height-30)
            a = im.crop(box)
            voter_img=im.crop((j+width-195,i+5,j+width-15,i+40))
            voter_list.append(voter_img)
            black = (0,0,0)
            white = (255,255,255)
            threshold = (202,202,202)

            img = a.convert("LA")
            pixels = img.getdata()

            newPixels = []

            for pixel in pixels:
              if pixel < threshold:
                newPixels.append(black)
              else:
                newPixels.append(white)

            newImg = Image.new("RGB",img.size)
            newImg.putdata(newPixels)
            list.append(newImg)
            
            # a.save("geeks{}.jpg".format(k))
            # list.append(a)
            # k=k+1
    return (list,voter_list)       
```
